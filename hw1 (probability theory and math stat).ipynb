Задача 1. Из колоды в 52 карты вынимают случайным образом 4 карты. Найти число исходов, соответствующих тому, что был вытянут хотя бы один туз.

4 карты из колоды можно вытянуть С(4,52) способами. 

Возможны четыре ситуации:
1. Один туз, все остальные карты - нет
2. Два туза.
3. Три туза.
4. Четыре туза.
События независимы и произойти может только одно из четырёх.
В колоде 4 туза. Один из четрыхёх можно выбрать С(1,4) способами, остальные карты С(3,48)
Для остальных способов размышления аналогичны.
Далее число способов можно просуммировать.

```python
import numpy as np

def permutations(n: int) -> int:
   
    return np.math.factorial(n)

def perm(n: int) -> int:
   
    return permutations(n)

def combinations(n: int, k: int) -> int:
    return np.math.factorial(n) // (np.math.factorial(k)*np.math.factorial(n-k))

def comb(n: int, k: int) -> int:
    return combinations(n, k)
 

def arrangement(n: int, k: int) -> int:
    return combinations(n,k)*perm(k)

def arrang(n: int, k: int) -> int:
    return arrangement(n, k)
```

```python
a1 = comb(4, 1)*comb(48, 3)
a2 = comb(4, 2)*comb(48, 2)
a3 = comb(4, 3)*comb(48, 1)
a4 = comb(4, 4)*comb(48, 0)
print(a1 + a2 + a3 + a4)
```

Задача 5

Допустим, имеется некоторая очень редкая болезнь (поражает 0.1% населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет 99% больных этой болезнью и всего лишь в 1% случаев даёт ложный положительный ответ.
Вопрос: какова вероятность, что вы действительно больны ей?

Подсказка: вновь используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности.

Гипотезы: 
B1 - человек болен
B2 - человек здоров

События:
A - прибор показал, что человек болен


Необходимо определить вероятность того, что имела место гипотеза "человек болен" при выполнении события А

Для нахождения полной вероятности события A необходимо посчитать полную вероятность этого события

Полная вероятность события A равна P(B1)*P(A|B1) + P(B2)* P(A|B2)
Зная значение вероятности события P(A|B1) = 0.99 можно вычислить вероятность гипотезы B1 через уравнение Байеса


```python
PB1 =  0.001
PB2 = 1 - PB1

P = PB1 * 0.99 + PB2 * 0.01
print("Вероятность, что человек на самом деле болен, равна {:f}".format(PB1*PB1*0.99/P))
```
