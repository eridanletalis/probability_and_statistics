
    Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.
    



```python
import numpy as np

a = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])

mean = a.sum()/a.size

print(f"Рассчитанное значение среднего арифметического \"вручную\" {mean:.4f}")

print(f"Рассчитанное значение среднего арифметического функцией mean {a.mean():.4f}")
```

Смещённая и несмещённая оценки дисперсии 
Дисперсия - мера разброса СВ относительно её математического ожидания. Единица измерения - квадрат той величины, дисперсию которой расчитываем. Квадратный корень дисперсии - СКО 
Дисперсия рассчитывается, как математическое ожидание квадратов расстояний между СВ и её математическим ожиданием. 
Выборочная дисперсия - оценка теоретической дисперсии распределения, рассчитанная на основе данных выборки. Она является смещённой оценкой теоретической дисперсии. 
Исправленная выборочная дисперсия - несмещённой. Несмещённую оценку дисперсии можно получить, домножив смещённую на (n-1)/n, где n - число элементов выборки. Или заменив знаменатель в формуле на (n - 1).
Считается, что исправленная выборочная дисперсия показывает оценку дисперсии для всех вариантов генеральной совокупности.

```python
D = ((a - mean) ** 2).sum()/a.size # смещённая дисперсия

print(f"Выборочная смещённая дисперсия, рассчитанная \"вручную\" {D:.4f}")

S2 = ((a - mean) ** 2).sum()/ (a.size - 1) # исправленная дисперсия

print(f"Исправленная выборочная дисперсия, рассчитанная \"вручную\" {S2:.4f}")
```

В больших выборках относительная разница между смещённой и несмещённой дисперсией практически отсутствует. Математически можно доказать, что предел ряда n/(n-1) при росте n стремится к единице, а с точки зрения здравого смысла - чем больше опытов мы проводим, тем ближе наша оценка к оценке генеральной совокумности.

Среднее квадратичное отклонение (СКО, стандартное отклонение) - квадратный корень из дисперсии. Показывает рассеяние СВ относительно её математического ожидания.

```python
sigma = S2 ** 0.5

print(f"Оценка СКО, рассчитанная \"вручную\" {sigma:.4f}")
print(f"Оценка СКО, c использованием функции std с параметром ddof=1 равна {a.std(ddof=1):.4f}")
```

В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?
  

```python
def C(n: int, k: int) -> int:
    
    return np.math.factorial(n) // (np.math.factorial(k)*np.math.factorial(n-k))

"""
Нас интересуют следующие случаи (знак | разделяет ящики):

ББ | БЧЧЧ C(5,2)/C(8,2) * C(5,1)С(7,3)/C(12,4)
БЧ | ББЧЧ C(5,1)C(3,1)/C(8,2) * C(5,2)C(7,2)/C(12,4)
ЧЧ | БББЧ C(3,2)/C(8,2) * C(5,3)C(7,1)/C(12,4)


Ситуации не зависимые, будем использовать сложение вероятностей
"""

p1 = (C(5,2) / C(8,2)) * (C(5,1) * C(7,3) / C(12,4))
p2 = (C(5,1) * C(3,1) / C(8,2)) * (C(5,2) * C(7,2) / C(12,4))
p3 = (C(3,2) / C(8,2)) * (C(5,3) * C(7,1) / C(12,4))

print(f"Вероятность того, что 3 мяча белые, равна {p1 + p2 + p3:.4f}")
```

  На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.
  

```python
"""
Имеем классическую задачу на условную вероятность.
Пусть вероятность попадания для 1 есть P_A, для второго P_B, для третьего P_C.
Событие, означающее попадание в мишень, обозначим H. Его вероятность (полная вероятность) - P_H

Таким образом, требуется найти P(A|H), P(B|H), P(C|H)
Для расчёта будет использована формула P(A|H) = P_A * (1/3) / P_H, где (1/3) - вероятность того, что
выбран стрелок A. 
"""
P = 1 / 3
P_A, P_B, P_C = [0.9, 0.8, 0.6]

P_H = (P_A + P_B + P_C) / 3 # Выбор каждого спортсмена равновозможен
P_A_H = P_A / P_H / 3
P_B_H = P_B / P_H / 3
P_C_H = P_C / P_H / 3
print(f"Вероятность того, что успешный выстрел произведён стрелком А равна {P_A_H:.4f}")
print(f"Вероятность того, что успешный выстрел произведён стрелком B равна {P_B_H:.4f}")
print(f"Вероятность того, что успешный выстрел произведён стрелком C равна {P_C_H:.4f}")
print(f"Контроль. 1.0 = {P_A_H + P_B_H + P_C_H}")
```

  В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?
  

Рассуждения аналогичны предыдущей задаче
Вероятность того, что будет выбран студент из группы А - 1/4, B - 1/4, С - 1/2. (x+x+2x = 4x; x/4x; x/4x; 2x/4x) Событие H - студент сдал первую сессию. Посчитаем полную вероятность события H - выбранный студент сдал первую сессию: 

P_H = 0.25 * 0.8 + 0.25 * 0.7 + 0.5 * 0.9


```python
P_H = 0.25 * 0.8 + 0.25 * 0.7 + 0.5 * 0.9
P_A_H = 0.25 * 0.8 / P_H
P_B_H = 0.25 * 0.7 / P_H
P_C_H = 0.5 * 0.9 / P_H


print(f"Вероятность того, студент учится на факультете А равна {P_A_H:.4f}")
print(f"Вероятность того, студент учится на факультете B равна {P_B_H:.4f}")
print(f"Вероятность того, студент учится на факультете C равна {P_C_H:.4f}")
print(f"Контроль. 1.0 = {P_A_H + P_B_H + P_C_H}")
```

  Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

```python
print(f"Вероятность выхода из строя всех деталей равна {0.1 * 0.2 * 0.25:.4f}")
p1 = 0.1; q1 = 1 - p1
p2 = 0.2; q2 = 1 - p2
p3 = 0.25; q3 = 1 - p3
print("Вероятность выхода из строя только двух деталей равна {:.4f}".format( \
p1 * p2 * q3 + p1 * q2 * p3 + q1 * p2 * p3))

print(f"Вероятность выходя из строя хотя бы одной детали равна {1 - q1 * q2 * q3:.4f}")

# Интересный резульатат. Если предположить, что вероятность одновременного выхода из строя двух какиих-то 
# деталей равна произведению вероятностей, то, воспользовавшись формулой вероятности совместных событий
print("Вероятность выходя из строя хотя бы одной детали, ИЛИ 1 ИЛИ 2 ИЛИ 3, равна {:.4f} (ВНЕ ЗАЧЁТА)".format(\
p1 + p2 + p3 - p1 * p2 - p2 * p3 - p1 * p2 * p3)) # Результат будет очень близкий. Почему?

# В чётвёртом случае можно воспользоваться результатми третьего, когда вышла из строя хотя бы одна деталь.
# так мы убираем вероятности того, что ни одна деталь из строя не вышла. И ограничиваем сверху
# вероятностью того, что из строя вышли все три детали
print(f"Вероятность, что выйдут из строя от одной до двух деталей, равна {1-q1*q2 * q3 - p1 * p2 * p3:.4f}")
```
