Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
Найдите ее среднее значение и дисперсию.

```python
import numpy as np

a = 200
b = 800

M = (a + b) / 2
print(f"Среднее значение равно {M:.2f}")
D = ((b - a) ** 2) / 12
print(f"Дисперсия равна {D: .2f}")
```

О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
Если да, найдите ее.

Математическое ожидание $$M(X) = \frac {a + b}{2}$$
Дисперсия $$D(X) = \frac {(b - a) ^ 2} {12}$$

Из последней формулы находим правую границу: $$b = \sqrt{12D} + a$$
Затем подставляем в формулы мат ожидания.

```python
a = 0.5
D = 0.2

b = np.sqrt(12 * D) + a
M = (a + b) / 2

print(f"Среднее значение равно {M:.2f}")
D = ((b - a) ** 2) / 12
print(f"Правая граница равна {b: .2f}")
```

Рост взрослого населения города X имеет нормальное распределение.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
    
а). больше 182 см
б). больше 190 см
в). от 166 см до 190 см
г). от 166 см до 182 см
д). от 158 см до 190 см
е). не выше 150 см или не ниже 190 см
ё). не выше 150 см или не ниже 198 см
ж). ниже 166 см.

1. Найдём значения сигм

Интервал $\sigma = 0.68$, Интервал для $2\sigma$ без первой и третей $0.274$. Интервал $3\sigma$ без остальных $0.0432‬$. За тремя сигмами интевал равен $0.0028‬$

$$\sigma = 8, M(X) = 174 $$

$M + \sigma = 182$ - a), первая сигма

$M + 2\sigma = 190$ - б), вторая сигма

$M - \sigma = 166$ - в), первая сигма

$M - 2\sigma = 158$ - г), вторая сигма

$M - 3\sigma = 150$ - е), третья сигма

$M + 3\sigma = 198$ - ё), третья сигма


2. Посчитаем вероятности

а) больше 182. Это значение - 1 первая сигма. То, что лежит за пределами первой сигмы может появится с вероятностью $1 - 0.68 = 0.32$, учитывая симметричность распределения, 0.32 делим пополам. Итого 0.16

б) больше 190. Аналогично, но 2 сигма. $P = \frac {1 - 0.954} {2} = 0.023$

в) $P(166 < X < 190) = P(X>166) + P(X<190) = 0.68 + \frac{0.274}{2} = 0.817$

 $\uparrow$ Вероятность в пределхах первой сигмы, плюс половина второй.

г) $P(166 < X < 190) = P(X>166) + P(X<182) = 2* P(X>M-\sigma) = 0.68$

 $\uparrow$ Вероятность в пределах первой сигмы

д) $P(158 < X < 190) = P(X>152) + P(X<190) = P(M - 2\sigma < X < M + 2\sigma) = 0.954$


е) $P(X<150) + P(X>190) = (1-0.9972) + \frac {1 - 0.954}{2}  = 0.0258‬$

 $\uparrow$ Всё, что за пределами трёх сигм, плюс половина третьей


ё) $P(X<150) + P(X>198) = 1 - 0.9972 = 0.0028$


ж) $P(X<166) = \frac {1 - 0.68}{2} =0.16 $

$\uparrow$ Всё, что от первой сигмы и меньше


На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от
математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

```python
M = 178
D = 25
sigma = np.sqrt(D)
X = 190
```

```python
(X - M) / sigma
```
